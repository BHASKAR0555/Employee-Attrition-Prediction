```python
data = pd.read_csv('train.csv')
tests = pd.read_csv('test.csv')
```

```python
data_norm = (data - data.mean()) / (data.max() - data.min())
data_norm.head()
tests_norm=(tests-tests.mean()) / (tests.max()- tests.min())
tests_norm.head()
```

```python
field={'Medical': 1,'Life Sciences': 2,'Other': 3,'Marketing': 4,'Technical Degree': 5,'Human Resources': 6}
travel={'Non-Travel':1, 'Travel_Rarely':2, 'Travel_Frequently':3}
dept={'Research & Development':1, 'Sales':2, 'Human Resources':3}
gender = {'Male': 1,'Female': 2} 
job={'Laboratory Technician':1, 'Manufacturing Director':2,
       'Sales Executive':3, 'Research Scientist':4, 'Sales Representative':5,
       'Healthcare Representative':6, 'Research Director':7,
       'Human Resources':8, 'Manager':9}
overtime = {'No': 1,'Yes': 2} 
marital={'Single':1, 'Married':2, 'Divorced':3}
```

```python
data.EducationField = [field[item] for item in data.EducationField]
data.BusinessTravel = [travel[item] for item in data.BusinessTravel]
data.Department = [dept[item] for item in data.Department] 
data.OverTime = [overtime[item] for item in data.OverTime]
data.MaritalStatus = [marital[item] for item in data.MaritalStatus]
data.JobRole = [job[item] for item in data.JobRole]
data.Gender = [gender[item] for item in data.Gender]
tests.EducationField = [field[item] for item in tests.EducationField]
tests.BusinessTravel = [travel[item] for item in tests.BusinessTravel]
tests.Department = [dept[item] for item in tests.Department] 
tests.OverTime = [overtime[item] for item in tests.OverTime]
tests.MaritalStatus = [marital[item] for item in tests.MaritalStatus]
tests.JobRole = [job[item] for item in tests.JobRole]
tests.Gender = [gender[item] for item in tests.Gender]
```

```python
data = data.drop(['Id', 'EmployeeNumber','Behaviour'], axis=1)
tests = tests.drop(['Id', 'EmployeeNumber','Behaviour'], axis=1)
```

```python
# Create correlation matrix
corr_matrix = data_norm.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.77)]

# Drop features 
data_norm.drop(to_drop, axis=1, inplace=True)
tests_norm.drop(to_drop, axis=1, inplace=True)
```
